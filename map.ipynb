```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных из файла churn.csv
df = pd.read_csv('C:/Users/user/Desktop/churn.csv')

# Разделение данных по признаку оттока
churned_customers = df[df['Exited'] == 1]
loyal_customers = df[df['Exited'] == 0]

# Построение scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(churned_customers['CreditScore'], churned_customers['EstimatedSalary'], color='red', label='Ушедшие клиенты', alpha=0.5)
plt.scatter(loyal_customers['CreditScore'], loyal_customers['EstimatedSalary'], color='blue', label='Лояльные клиенты', alpha=0.5)

plt.title('Взаимосвязь кредитного рейтинга и предполагаемой зарплаты')
plt.xlabel('Кредитный рейтинг')
plt.ylabel('Предполагаемая зарплата')
plt.legend()
plt.grid(True)
plt.show()
#На графике есть явная зависимость между кредитным рейтингом и предполагаемой зарплатой, мы увидим, что точки как-то структурированы, возможно, образуя какой-то общий тренд
#Однако, если точки распределены хаотично без какой-либо видимой закономерности, это может указывать на отсутствие явной взаимосвязи между этими признаками.
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd

# Загрузка данных о странах и границах мировых стран
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Загрузка данных о доле ушедших клиентов по странам
churn_data = pd.read_csv('C:/Users/user/Desktop/churn.csv')

# Группировка данных по странам и вычисление средней доли ушедших клиентов в каждой стране
country_churn_rate = churn_data.groupby('Geography')['Exited'].mean().reset_index()

# Объединение данных о странах с данными о доле ушедших клиентов
world = world.merge(country_churn_rate, how='left', left_on='name', right_on='Geography')

# Построение тепловой картограммы
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
world.plot(column='Exited', cmap='Reds', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)
plt.title('Доля ушедших клиентов по странам')
plt.xlabel('Долгота')
plt.ylabel('Широта')
ax.set_xlim(-150, 150)
ax.set_ylim(-90, 90)
world.boundary.plot(ax=ax, linewidth=1.0)
plt.show()
```
