```python
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd

#        
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

#        
churn_data = pd.read_csv('C:/Users/user/Desktop/churn.csv')

#             
country_churn_rate = churn_data.groupby('Geography')['Exited'].mean().reset_index()

#          
world = world.merge(country_churn_rate, how='left', left_on='name', right_on='Geography')

#   
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
world.plot(column='Exited', cmap='Reds', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)
plt.title('    ')
plt.xlabel('')
plt.ylabel('')
ax.set_xlim(-150, 150)
ax.set_ylim(-90, 90)
world.boundary.plot(ax=ax, linewidth=1.0)
plt.show()
```

```python
import pandas as pd

#        
def get_credit_score_cat(credit_score):
    if credit_score >= 300 and credit_score < 500:
        return "Very_Poor"
    elif credit_score >= 500 and credit_score < 601:
        return "Poor"
    elif credit_score >= 601 and credit_score < 661:
        return "Fair"
    elif credit_score >= 661 and credit_score < 781:
        return "Good"
    elif credit_score >= 781 and credit_score < 851:
        return "Excellent"
    elif credit_score >= 851:
        return "Top"
    elif credit_score < 300:
        return "Deep"

#     CreditScore     CreditScoreCat
churn_data['CreditScoreCat'] = churn_data['CreditScore'].apply(get_credit_score_cat)

#   
pivot_table = pd.pivot_table(churn_data, 
                             values='Exited', 
                             index='CreditScoreCat', 
                             columns='Tenure', 
                             aggfunc='mean')

#     
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt=".2%")
plt.title('         ')
plt.xlabel(' ')
plt.ylabel('  ')
plt.show()
```
